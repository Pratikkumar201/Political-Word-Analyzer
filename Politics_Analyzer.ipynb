```python
import pandas as pd
import numpy as np
import fitz  # PyMuPDF
from sklearn.feature_extraction.text import TfidfVectorizer
from google.colab import files
import os

# Step 1: Upload dictionary
print("📁 Upload your dictionary CSV (15 or 20 clusters)")
dict_file = files.upload()
dict_path = list(dict_file.keys())[0]

df_dict = pd.read_csv(dict_path)
word_list = []
for col in df_dict.columns:
    word_list += df_dict[col].dropna().astype(str).str.strip().tolist()
word_list = list(set(word_list))  # Unique words only

# Step 2: Upload all 23 speeches (PDF)
print("📂 Upload all 23 speeches (named like A1.pdf, N2.pdf, P4.pdf, Y7.pdf)")
uploaded = files.upload()

# Step 3: Extract text from each PDF
speech_texts = []
file_codes = []

def extract_text(path):
    text = ""
    with fitz.open(path) as doc:
        for page in doc:
            text += page.get_text()
    return text

for filename in uploaded:
    code = os.path.splitext(filename)[0].strip().replace(" ", "_")
    text = extract_text(filename)
    speech_texts.append(text)
    file_codes.append(code)

# Step 4: TF-IDF Vectorization
def identity_tokenizer(text):
    return text.split()

vectorizer = TfidfVectorizer(
    vocabulary=word_list,
    tokenizer=identity_tokenizer,
    lowercase=False,
    token_pattern=None
)
X = vectorizer.fit_transform(speech_texts)
words = vectorizer.get_feature_names_out()
idf = vectorizer.idf_

# Step 5: Build wide TF-IDF table
final_df = pd.DataFrame({'Word': words})

for i, code in enumerate(file_codes):
    tf = X[i].toarray().flatten()
    final_df[f'{code}_TF'] = tf
    final_df[f'{code}_IDF'] = idf  # same IDF for each speech

# Step 6: Save and Download
output_file = "TFIDF_23_Speeches_Wide_Table.csv"
final_df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"✅ Table saved as: {output_file}")
files.download(output_file)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import fitz  # PyMuPDF
from google.colab import files
import os

# STEP 1: Load dictionary CSV
dict_path = '/content/drive/MyDrive/Harsh Agrawal PhD/Category1_data.csv'
df_dict = pd.read_csv(dict_path)

# STEP 2: Flatten word list
word_list = []
for col in df_dict.columns:
    word_list.extend(df_dict[col].dropna().astype(str).str.strip().tolist())
word_list = list(set(word_list))

# STEP 3: Upload speech PDFs
print("📂 Upload all Phase 1 leader speeches (PDFs)")
uploaded = files.upload()

# STEP 4: Analyze speeches
leader_columns = {}

def identity_tokenizer(text):
    return text.split()

vectorizer = CountVectorizer(
    vocabulary=word_list,
    tokenizer=identity_tokenizer,
    lowercase=False,
    token_pattern=None
)

for filename in uploaded:
    leader_name = os.path.splitext(filename)[0].replace(" ", "_")

    # Extract text
    text = ""
    with fitz.open(filename) as doc:
        for page in doc:
            text += page.get_text()

    # CountVectorizer
    X = vectorizer.fit_transform([text])
    tf_counts = X.toarray().flatten()
    words = vectorizer.get_feature_names_out()

    # Convert to TF scores (normalized)
    total_terms = tf_counts.sum()
    tf_scores = tf_counts / total_terms if total_terms > 0 else tf_counts

    tf_df = pd.DataFrame({'Word': words, 'TF_Score': tf_scores})
    tf_df = tf_df[tf_df['TF_Score'] > 0]  # used words only

    # Top 50% TF Score words
    threshold = tf_df['TF_Score'].median()
    top_df = tf_df[tf_df['TF_Score'] >= threshold].copy()
    top_df = top_df.sort_values(by='TF_Score', ascending=False)

    # Format: WORD (TF Score)
    formatted = top_df.apply(lambda row: f"{row['Word']} ({round(row['TF_Score'], 4)})", axis=1).tolist()

    # Store in final dict
    leader_columns[leader_name] = formatted

# STEP 5: Build table
max_len = max(len(v) for v in leader_columns.values())
for k in leader_columns:
    leader_columns[k] += [""] * (max_len - len(leader_columns[k]))

final_df = pd.DataFrame(leader_columns)

# Show output
print("\n📋 Final Top 50% TF Score Table:")
display(final_df)

# # Save
# final_df.to_csv("top_50_percent_tf_score_table.csv", index=False, encoding='utf-8-sig')
# print("✅ Saved as: top_50_percent_tf_score_table.csv")

# Save
output_file = "top_50_percent_tf_score_table.csv"
final_df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"✅ Saved as: {output_file}")
files.download(output_file)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import fitz  # PyMuPDF
from google.colab import files
import os

# STEP 1: Upload Dictionary CSV
print("📁 Upload dictionary CSV (20-cluster or 15-cluster)")
dict_file = files.upload()
dict_path = list(dict_file.keys())[0]

# Load cluster dictionary
df_dict = pd.read_csv(dict_path)
word_cluster_pairs = []
for col in df_dict.columns:
    words = df_dict[col].dropna().astype(str).str.strip()
    word_cluster_pairs.extend([(word, col) for word in words if word])
word_cluster_df = pd.DataFrame(word_cluster_pairs, columns=["Word", "Cluster"])
word_list = word_cluster_df["Word"].unique().tolist()

# STEP 2: Upload Phase files
phase_name = input("Enter Phase name (e.g., Phase1): ").strip()
print(f"📂 Upload all leader speeches for {phase_name}")
uploaded = files.upload()

# STEP 3: Read and process each leader's file
def identity_tokenizer(text):
    return text.split()

leader_cluster_scores = {}

for filename in uploaded.keys():
    leader = os.path.splitext(filename)[0].replace(" ", "_")

    # Extract text
    text = ""
    with fitz.open(filename) as doc:
        for page in doc:
            text += page.get_text()

    # TF-IDF
    vectorizer = TfidfVectorizer(
        vocabulary=word_list,
        tokenizer=identity_tokenizer,
        lowercase=False,
        token_pattern=None
    )
    X = vectorizer.fit_transform([text])
    words = vectorizer.get_feature_names_out()
    tf = X.toarray().flatten()

    tf_df = pd.DataFrame({"Word": words, "TF": tf})
    tf_df = tf_df[tf_df["TF"] > 0]

    merged_df = tf_df.merge(word_cluster_df, on="Word")
    cluster_scores = merged_df.groupby("Cluster")["TF"].sum().round(3)
    leader_cluster_scores[leader] = cluster_scores

# STEP 4: Create final DataFrame
result_df = pd.DataFrame(leader_cluster_scores).fillna(0).reset_index()
result_df = result_df.rename(columns={"index": "Cluster"}).sort_values("Cluster")

# STEP 5: Save CSV
csv_filename = f"{phase_name}_cluster_tf_table.csv"
result_df.to_csv(csv_filename, index=False)
print(f"✅ Table saved: {csv_filename}")

# STEP 6: Plot & Save Line Chart
plt.figure(figsize=(12, 8))
for leader in result_df.columns[1:]:
    plt.plot(result_df["Cluster"], result_df[leader], marker='o', label=leader)


    # Add value labels at each point
    for x, y in zip(result_df["Cluster"], result_df[leader]):
        if y != 0:
            plt.text(x, y + 0.03, f"{y:.3f}", ha='center', va='bottom', fontsize=12)
plt.title(f"{phase_name} - Cluster-wise TF Score")
plt.xlabel("Cluster")
plt.ylabel("TF Score")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()

chart_filename = f"{phase_name}_tf_chart.png"
plt.savefig(chart_filename)
plt.show()
print(f"📈 Chart saved: {chart_filename}")

# STEP 7: Download both files
files.download(csv_filename)
files.download(chart_filename)
```

```python
import pandas as pd
import numpy as np
import fitz  # PyMuPDF
from sklearn.feature_extraction.text import TfidfVectorizer
from google.colab import files
import os

# Step 1: Upload dictionary CSV
print("📁 Upload dictionary CSV (15 or 20 clusters)")
dict_file = files.upload()
dict_path = list(dict_file.keys())[0]

df_dict = pd.read_csv(dict_path)
word_list = []
for col in df_dict.columns:
    word_list += df_dict[col].dropna().astype(str).str.strip().tolist()
word_list = list(set(word_list))

# Step 2: Upload all 23 speeches
print("📂 Upload all 23 speeches (A1–A7, N1–N6, P1–P4, Y1–Y6)")
uploaded = files.upload()

# Step 3: Extract text per leader
def extract_text(file_path):
    text = ""
    with fitz.open(file_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

leader_map = {
    "A": "Akhilesh",
    "N": "Modi",
    "P": "Priyanka",
    "Y": "Yogi"
}

leader_texts = {name: [] for name in leader_map.values()}

for filename in uploaded:
    code = os.path.splitext(filename)[0].strip().upper()
    leader_key = code[0]
    leader = leader_map.get(leader_key)
    if leader:
        text = extract_text(filename)
        leader_texts[leader].append(text)

# Step 4: Global IDF fit
def identity_tokenizer(text):
    return text.split()

all_texts = sum(leader_texts.values(), [])
vectorizer = TfidfVectorizer(
    vocabulary=word_list,
    tokenizer=identity_tokenizer,
    lowercase=False,
    token_pattern=None
)
X = vectorizer.fit_transform(all_texts)
idf_scores = vectorizer.idf_
idf_df = pd.DataFrame({'Word': vectorizer.get_feature_names_out(), 'IDF': idf_scores})

# Step 5: Top/Bottom 25% per leader (unique only)
bottom_dict = {}
top_dict = {}

# Keep track of all words assigned to avoid duplicates
bottom_used = set()
top_used = set()

for leader, speeches in leader_texts.items():
    combined = " ".join(speeches)
    X_leader = vectorizer.transform([combined])
    tfidf_array = X_leader.toarray().flatten()
    words = vectorizer.get_feature_names_out()
    used_words = words[tfidf_array > 0]

    df = pd.DataFrame({'Word': used_words})
    df = df.merge(idf_df, on='Word')

    # Remove already used words by others (for uniqueness)
    bottom_25 = df.nsmallest(len(df) // 4, "IDF")
    top_25 = df.nlargest(len(df) // 4, "IDF")

    bottom_unique = bottom_25[~bottom_25['Word'].isin(bottom_used)]
    top_unique = top_25[~top_25['Word'].isin(top_used)]

    # Store used words for future filtering
    bottom_used.update(bottom_unique['Word'])
    top_used.update(top_unique['Word'])

    # Format with (IDF)
    bottom_dict[leader] = bottom_unique.apply(lambda r: f"{r['Word']} ({round(r['IDF'], 2)})", axis=1).tolist()
    top_dict[leader] = top_unique.apply(lambda r: f"{r['Word']} ({round(r['IDF'], 2)})", axis=1).tolist()

# Step 6: Format as wide tables
def format_wide_table(data_dict):
    max_len = max(len(v) for v in data_dict.values())
    for k in data_dict:
        data_dict[k] += [""] * (max_len - len(data_dict[k]))
    return pd.DataFrame(data_dict)

bottom_df = format_wide_table(bottom_dict)
top_df = format_wide_table(top_dict)

# Step 7: Save and Download
bottom_file = "Q3_Unique_Bottom_25_IDF.csv"
top_file = "Q3_Unique_Top_25_IDF.csv"

bottom_df.to_csv(bottom_file, index=False, encoding='utf-8-sig')
top_df.to_csv(top_file, index=False, encoding='utf-8-sig')

print("✅ Unique top & bottom 25% IDF tables saved.")
files.download(bottom_file)
files.download(top_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# STEP 1: Upload all 4 leader IDF CSVs
print("📁 Upload all 4 leader-wise avg_idf_per_cluster CSVs")
uploaded = files.upload()

# STEP 2: Read and merge them into one DataFrame
combined_df = None

for file in uploaded:
    leader = file.replace("_avg_idf_per_cluster.csv", "").strip()
    df = pd.read_csv(file)
    df = df.rename(columns={"IDF": leader})

    if combined_df is None:
        combined_df = df
    else:
        combined_df = pd.merge(combined_df, df, on="Cluster", how="outer")

combined_df = combined_df.sort_values("Cluster").fillna(0)

# STEP 3: Plot line chart
plt.figure(figsize=(12, 6))
colors = plt.cm.get_cmap('tab10', len(combined_df.columns[1:]))

for i, leader in enumerate(combined_df.columns[1:]):
    color = colors(i)
    plt.plot(combined_df["Cluster"], combined_df[leader], marker='o', label=leader, color=color)

    # Add value labels
    for x, y in zip(combined_df["Cluster"], combined_df[leader]):
        if y > 0:
            shift = 0.02 * ((i % 3) - 1)  # alternate shift to prevent overlap
            plt.text(x, y + shift, f"{y:.2f}", ha='center', fontsize=8, color=color)

plt.title("Leader-wise Average IDF per Cluster")
plt.xlabel("Cluster")
plt.ylabel("Average IDF")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()

# STEP 4: Save and Download
chart_file = "AllLeaders_Avg_IDF_LineChart.png"
plt.savefig(chart_file)
plt.show()
files.download(chart_file)

# Optional: Save merged table too
combined_df.to_csv("AllLeaders_Avg_IDF_Table.csv", index=False)
files.download("AllLeaders_Avg_IDF_Table.csv")
```

```python
import pandas as pd
import plotly.graph_objs as go
from google.colab import files

# STEP 1: Upload all leader-wise IDF CSVs
print("📁 Upload all leader-wise avg_idf_per_cluster CSVs")
uploaded = files.upload()

# STEP 2: Combine CSVs into a single DataFrame
combined_df = None

for file in uploaded:
    leader = file.replace("_avg_idf_per_cluster.csv", "").strip()
    df = pd.read_csv(file)
    df = df.rename(columns={"IDF": leader})

    if combined_df is None:
        combined_df = df
    else:
        combined_df = pd.merge(combined_df, df, on="Cluster", how="outer")

combined_df = combined_df.sort_values("Cluster").fillna(0)

# STEP 3: Create interactive Plotly line chart
fig = go.Figure()

for leader in combined_df.columns[1:]:
    fig.add_trace(go.Scatter(
        x=combined_df["Cluster"],
        y=combined_df[leader],
        mode='lines+markers+text',
        name=leader,
        text=[f"{v:.2f}" if v > 0 else "" for v in combined_df[leader]],
        textposition="top center"
    ))

fig.update_layout(
    title="📊 Leader-wise Average IDF per Cluster",
    xaxis_title="Cluster",
    yaxis_title="Average IDF",
    xaxis=dict(tickangle=45),
    template="plotly_white",
    hovermode="x unified",
    legend=dict(title="Leaders")
)

fig.show()

# STEP 4: Save as HTML (optional)
html_file = "Leaderwise_Avg_IDF_Dashboard.html"
fig.write_html(html_file)
files.download(html_file)

# Save merged table as well
combined_df.to_csv("AllLeaders_Avg_IDF_Table.csv", index=False)
files.download("AllLeaders_Avg_IDF_Table.csv")
```
